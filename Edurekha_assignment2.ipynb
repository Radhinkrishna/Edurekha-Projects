```python
import pandas as pd
df=pd.read_csv('/content/top50spotify.csv')
df.head()
```

```python
df.columns
```

```python
#2
df.to_csv('top50.csv')
#3
f_10=df.head(10)
avg_engy=f_10['Energy'].mean()
avg_ln=f_10['Length.'].mean()
print(f'Average Energy: {avg_engy}')
print(f'Average Length: {avg_ln}')
```

```python
#4
df4=df.groupby('Genre')['Length.'].mean()
print(df4)
```

```python
#5
df5 = df.groupby(['Artist.Name', 'Genre']).size()
mst=df5.sort_values(ascending=False).head(1)
#6
print("Artist  with the most number of tracks in one genre\n",mst)
print("\n \n \n",df5)
```

#problem 2
1.Create a pandasseries from the below dictionary where indices are subjects:
2.Convert the created series into DataFrame and replace the null values with zeroes.
3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects

```python
import pandas as pd
data={'English':{'Sam':60,'Jackson':74,'Ahree':85},
              'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
      'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
      'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
      'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
#1
pb2sr=pd.Series(data)
print(pb2sr)
```

```python
#2
pb2df=pb2sr.to_frame()
pb2df.fillna(0,inplace=True)
print(pb2df)
#3
trns=pb2df.T
trns['Average']=trns.mean(axis=1)
print(trns)
```

```python
import pandas as pd
#1
pb4=pd.read_csv('/content/cereal.csv')
pb4.head()
```
